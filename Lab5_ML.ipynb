```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

def load_data():
    
    # Example dataset creation (replace this with your actual data)
    data = {
        'Feature': np.random.rand(100),  # Replace with your actual feature column
        'Target': np.random.rand(100)    # Replace with your actual target column
    }
    df = pd.DataFrame(data)
    return df

def split_data(df, test_size=0.2, random_state=42):

    X = df[['Feature']]  # Feature matrix
    y = df['Target']     # Target vector
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

def train_linear_regression(X_train, y_train):

    model = LinearRegression().fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    return model, y_train_pred

def plot_regression_results(X_train, y_train, y_train_pred):
    
    plt.scatter(X_train, y_train, color='blue', label='Actual')
    plt.plot(X_train, y_train_pred, color='red', label='Fitted line')
    plt.xlabel('Feature')
    plt.ylabel('Target')
    plt.legend()
    plt.show()

# Main program
if __name__ == "__main__":
    # Load the dataset
    df = load_data()
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_data(df)
    
    # Train the linear regression model
    model, y_train_pred = train_linear_regression(X_train, y_train)
    
    # Extract and display model coefficients
    slope = model.coef_
    intercept = model.intercept_
    print("Coefficient (slope):", slope)
    print("Intercept:", intercept)
    
    # Plot the actual vs predicted results
    plot_regression_results(X_train, y_train, y_train_pred)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

def evaluate_model(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    metrics = {
        'MSE': mse,
        'RMSE': rmse,
        'MAPE': mape,
        'R2': r2
    }
    return metrics

def display_metrics(metrics, dataset_name):
   
    print(f"Metrics for {dataset_name} set:")
    for key, value in metrics.items():
        print(f"{key}: {value:.4f}")
    print("\n")

# Main program
if __name__ == "__main__":
    # Load the dataset
    df = load_data()
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_data(df)
    
    # Train the linear regression model
    model, y_train_pred = train_linear_regression(X_train, y_train)
    
    # Evaluate the model on the training set
    train_metrics = evaluate_model(y_train, y_train_pred)
    display_metrics(train_metrics, "Train")
    
    # Perform prediction on the test data
    y_test_pred = model.predict(X_test)
    
    # Evaluate the model on the test set
    test_metrics = evaluate_model(y_test, y_test_pred)
    display_metrics(test_metrics, "Test")
    
    # Plot the actual vs predicted results for the training set
    plot_regression_results(X_train, y_train, y_train_pred)
    
    # Plot the actual vs predicted results for the test set
    plot_regression_results(X_test, y_test, y_test_pred)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

def load_and_split_data(file_path, test_size=0.2, random_state=42):
    
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Assume all columns except the last one are features, and the last column is the target
    X = df.iloc[:, :-1]  # Features (all columns except the last)
    y = df.iloc[:, -1]   # Target (last column)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    return X_train, X_test, y_train, y_test

def train_linear_regression(X_train, y_train):
    
    # Create and train the linear regression model
    model = LinearRegression().fit(X_train, y_train)
    
    # Predict on the training data
    y_train_pred = model.predict(X_train)
    
    return model, y_train_pred

def evaluate_model(y_true, y_pred):
    
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    metrics = {
        'MSE': mse,
        'RMSE': rmse,
        'MAPE': mape,
        'R2': r2
    }
    return metrics

def display_metrics(metrics, dataset_name):
    
    print(f"Metrics for {dataset_name} set:")
    for key, value in metrics.items():
        print(f"{key}: {value:.4f}")
    print("\n")

# Main program
if __name__ == "__main__":
    # Define the path to your dataset
    file_path = r"C:/Users/Admin/Documents/Students_data.csv"
    
    # Load and split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = load_and_split_data(file_path)
    
    # Train the linear regression model using all features
    model, y_train_pred = train_linear_regression(X_train, y_train)
    
    # Evaluate the model on the training set
    train_metrics = evaluate_model(y_train, y_train_pred)
    display_metrics(train_metrics, "Train")
    
    # Perform prediction on the test data
    y_test_pred = model.predict(X_test)
    
    # Evaluate the model on the test set
    test_metrics = evaluate_model(y_test, y_test_pred)
    display_metrics(test_metrics, "Test")
```
